```python
# Definition of function
def bin_votes(votes):
    if (votes >= 0) and (votes <= 999):
        tag = 'cat_1'
    elif (votes >= 1000) and (votes <= 9999):
        tag = 'cat_2'
    elif (votes >= 10000) and (votes <= 99999):
        tag = 'cat_3'
    elif (votes >= 100000) and (votes <= 999999):
        tag = 'cat_4'
    elif (votes >= 1000000) and (votes <= 9999999):
        tag = 'cat_5'
        
    return tag
```

```python
# Test the function 'bin'
bin_votes(1000)
```

```python
# Create new column called 'bin', and apply the function 'bin_votes' to column 'Votes'. Noted that 'axix=1' indicate that the 
# operation is by each column
df['bin'] = df.apply(lambda row: bin_votes(row['Votes']), axis=1)
df.head()
```

### Challenge 2

```python
# Obtain a new column to calculate the Revenue per minute by each film

# Direct way
#df['revenue_per_min'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']

# Using lambda function
df['revenue_per_min'] = df.apply(lambda row: row['Revenue (Millions)'] / row['Runtime (Minutes)'], axis=1)
df.head()
```
